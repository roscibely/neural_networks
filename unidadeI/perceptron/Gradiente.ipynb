## Método da Descida do Gradiente

Seja a expressão: $$ E = \sum_i ((x_i  w) - y_i)^2 $$

Queremos calcular a derivada de $E$ em relação a $w$.

```python
def mse(w, x, y):
    y_pred = x * w
    e = y_pred - y
    E = np.sum(e**2)
    return E
```



Minimizando $E$ pelo gradiente descendente

$$ w_{k+1} = w_k - \lambda \frac {\partial E}{\partial w} $$

Supondo que valor inicial ($k=0$) $w_0 = 0$, use learning rate $\lambda = 0.01$ para calcular o valor do novo $w_{100}$, ou seja, fazendo 100 atualizações de gradientes. Deve-se usar a função `mse` criada anteriormente.



```python
import numpy as np
import matplotlib.pyplot as plt

# Gerando os dados
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Definindo os hyperparameters
learning_rate = 0.01
num_iterations = 10
num_samples = len(x)
w = 0
errors=[]
ww=[]

# Descida do gradiente
for i in range(num_iterations):
      y_pred = x * w
      e = y_pred - y
      dw = (2/num_samples) * np.sum(e * x)
      w -= learning_rate * dw
      error = mse(w, x, y)
      errors.append(error)
      print(f'Iteração : {i} ------ MSE {error}')
```

```python
# Plot error function
plt.plot(range(num_iterations), errors)
plt.xlabel('Iterações')
plt.ylabel('E(w,x,y)')
plt.title('Função custo')
plt.show()
```
